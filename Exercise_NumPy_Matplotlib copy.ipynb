```python
# Exercise One
# Draw a line in a diagram from position (1, 3) to position (8,10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3,10]) 

plt.plot(xpoints, ypoints)
plt.title('Exercise One')

plt.show()
```

```python
# Exercise Two
# Draw 2 points in a diagram from position (1, 3) to position (8,10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3,10]) 

plt.plot(xpoints, ypoints, 'o')
plt.title('Exercise Two')

plt.show()
```

```python
# Exercise Three
# Draw a line in a diagram from position (1, 3) to (2, 8) 
# then to (6, 1) and finally to position (8,10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o')
plt.plot(xpoints, ypoints)

plt.title('Exercise Three - plot point and draw a line')

plt.show()
```

```python
# Exercise Five
# Creating a scatter plot

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('DAP_SampleData\data.csv')

df.plot(kind = "scatter", x = "Duration", y = "Calories")

plt.title('Exercise Five - scatter plot')

plt.show()
```

```python
# Exercise Six
# Creating a Histogram plot

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('DAP_SampleData\data.csv')

df["Duration"].plot(kind ='hist')

plt.title('Exercise Six - Histogram')

plt.show()
```
